```python
import pandas as pd
```

```python
train = pd.read_csv('/content/Task2/train.csv')
extra = pd.read_csv('/content/Task2/extra.csv')
```

```python
merged = pd.concat([train, extra], ignore_index=True)
```

```python
merged
```

```python
l = [[0,0,0,1], [0,0,0,1]]
labels = []
for i in l:
  txt = []
  for index, k in enumerate(i):
    if k>0:
      txt.append(str(index))
  labels.append(" ".join(txt))
```

```python
import glob
submission_set = glob.glob('/*')

[s.split('/')[-1] for s in submission_set]
```

```python
sample = int(len(merged) *.8)


train = merged[:sample]

test = merged[sample:]

train.to_csv('/content/Task2/train.csv', index = False)
```
