```python
# labels = ks["train"].features["label"].names
labels = sorted(set(df_600['label']))
label2id, id2label = dict(), dict()
for i, label in enumerate(labels):
    label2id[label] = str(i)
    id2label[str(i)] = label
```

```python
label2id
```

```python
from transformers import AutoFeatureExtractor

feature_extractor = AutoFeatureExtractor.from_pretrained("facebook/wav2vec2-base")
```

```python
feature_extractor.sampling_rate
```

```python
def preprocess_function(examples):
    # print(examples["audio"])
    audio_arrays = [x[0]["array"] for x in examples["audio"]]
    # print(audio_arrays)
    inputs = feature_extractor(
        audio_arrays, sampling_rate=16000, max_length=3906, truncation=True
    )
    return inputs
```

```python
encoded_val = val_data.map(preprocess_function, remove_columns=["audio", "file"], batched=True)
encoded_train = train_data.map(preprocess_function, remove_columns=["audio", "file"], batched=True)
```

```python
from transformers import AutoModelForAudioClassification, TrainingArguments, Trainer

num_labels = len(id2label)
model = AutoModelForAudioClassification.from_pretrained(
    "facebook/wav2vec2-base", num_labels=num_labels, label2id=label2id, id2label=id2label
)
```

```python
import numpy as np
from datasets import load_metric

metric = load_metric("accuracy")
def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    return metric.compute(predictions=predictions, references=labels)
# def acc_and_f1(preds, labels):
#     acc = simple_accuracy(preds, labels)
#     f1 = f1_score(y_true=labels, y_pred=preds).item()
#     return {
#         "accuracy": acc,
#         "f1": f1,
#     }
```

```python
import transformers
```

```python
callbacks = transformers.EarlyStoppingCallback( early_stopping_patience=30, early_stopping_threshold= 0.0001)
```

```python
training_args = TrainingArguments(
    output_dir="./v1_600",
    evaluation_strategy="epoch",
    save_strategy="epoch",
    learning_rate=3e-5,
    num_train_epochs=500,
    per_device_train_batch_size=512,
    per_device_eval_batch_size=512,
    load_best_model_at_end=True,
    overwrite_output_dir=True,
    save_total_limit=3,
    warmup_steps=2000,
    logging_steps=2000,
    weight_decay = 0.0001
    # eval_steps=10,

)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=encoded_train,
    eval_dataset=encoded_val,
    tokenizer=feature_extractor,
    compute_metrics = compute_metrics
)

trainer.train()
```
